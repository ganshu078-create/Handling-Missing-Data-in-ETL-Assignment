#Handling Missing Data in ETL Assignment

#SECTION A – THEORETICAL QUESTIONS

**Q1. What are the most common reasons for missing data in ETL pipelines?**

- Missing data in ETL pipelines can occur due to several common reasons:

Human Error

Data may be left blank or entered incorrectly during manual input.

System or Technical Failures

Errors during data collection, transmission, or storage can result in missing values.

Data Integration Issues

When combining data from multiple sources, some fields may not match or may be unavailable.

Optional Fields

Some data fields are not mandatory, so users may skip filling them.

Data Corruption or Loss

Data can be lost due to file corruption, network issues, or processing errors.

**Q2. Why is blindly deleting rows with missing values considered a bad practice in ETL?**

- Blindly deleting rows with missing values is considered a bad practice because it can negatively impact the quality and usefulness of the dataset.

Reasons:

Loss of Important Data

Deleting rows may remove valuable information that could be useful for analysis.

Reduced Dataset Size

Excessive deletion can significantly shrink the dataset, making results less reliable.

Introduction of Bias

If missing values are not random, removing those rows can distort the data and lead to biased conclusions.

Loss of Patterns and Insights

Missing data itself can contain meaningful information, which gets lost when rows are deleted.

**Q3. Explain the difference between:**

**Listwise deletion**

**Column deletion**

Also mention one scenario where each is appropriate.

- Listwise Deletion:

Listwise Deletion means removing entire rows (records) that contain any missing values.

Key Points:

Deletes complete records
Used when missing data is very small
Simple and easy to apply

Example:
If one row has a missing value in any column, that entire row is removed.

- Column Deletion:

Column Deletion means removing entire columns (features) that contain a large number of missing values.

Key Points:

Deletes a whole column
Used when most values in that column are missing
Helps simplify the dataset

Example:
If a column like "Income" has too many missing values, it can be removed completely.

| Basis           | Listwise Deletion    | Column Deletion               |
| --------------- | -------------------- | ----------------------------- |
| What is removed | Rows (records)       | Columns (features)            |
| Use case        | Few missing values   | Many missing values in column |
| Impact          | Loss of some records | Loss of one feature entirely  |

**Q4. Why is median imputation preferred over mean imputation for skewed data such as income?**

- Median imputation is preferred over mean imputation for skewed data (like income) because it provides a more accurate representation of the dataset.

Reasons:

Less Affected by Outliers

Income data often contains extreme values (very high or very low incomes).

The mean gets influenced by these outliers, while the median remains stable.
Better Representation of Central Value
The median represents the middle value of the dataset, making it more reliable for skewed distributions.

Maintains Data Distribution

Using the median helps preserve the natural shape of the data without distortion.

**Q5. What is forward fill and in what type of dataset is it most useful?**

- Forward Fill (also called forward propagation) is a technique used to handle missing values by replacing them with the previous available (non-missing) value in the dataset.

How it works:

If a value is missing, it takes the last known value above it.

Example:

| Day | Sales |
| --- | ----- |
| 1   | 100   |
| 2   | NaN   |
| 3   | NaN   |
| 4   | 150   |

After Forward Fill:

| Day | Sales |
| --- | ----- |
| 1   | 100   |
| 2   | 100   |
| 3   | 100   |
| 4   | 150   |

Forward Fill is an effective method for handling missing values in time-based or ordered datasets where past values can reasonably represent missing ones.

**Q6. Why should flagging missing values be done before imputation in an ETL workflow?**

- Flagging missing values before imputation is important because it helps preserve information about the original data and improves analysis.

Reasons:

Retains Original Information

Once missing values are filled, you lose track of which values were originally missing. A flag column preserves this information.

Improves Data Analysis

Missing values themselves can carry meaning (e.g., non-response or special customer behavior). Flagging helps analyze this.

Better Model Performance

In machine learning, models can use the flag to understand patterns related to missing data.

Avoids Misleading Data Interpretation

Imputed values may look real, but they are estimates. Flagging helps differentiate between actual and filled data.

**Q7. Consider a scenario where income is missing for many customers. How can this missingness itself provide business insights?**

- Missing income data can itself provide useful business insights rather than being just a data problem.

Key Insights:

- Customer Privacy Concerns

Many customers may choose not to disclose income due to privacy reasons. This indicates a need to improve trust and data transparency.

- Customer Segmentation

Missing income may be common in certain groups (e.g., students, freelancers, or new users), helping businesses identify different customer segments.

- Data Collection Issues

A high number of missing values may indicate problems in forms, surveys, or data entry processes.

- Alternative Analysis Opportunities

Businesses can use other variables (like spending behavior or purchase history) to estimate customer value instead of relying only on income.

- Detection of Bias

If missing income is not random, it may lead to biased analysis. Identifying this helps improve decision-making.
